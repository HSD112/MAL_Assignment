# Keras - circles

In this notebook, we install Keras and create a simple, fully-connected network.

```python
%matplotlib inline

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.metrics import Accuracy
import mglearn

import matplotlib.pyplot as plt
import numpy as np

from sklearn.datasets.samples_generator import make_circles
from sklearn.model_selection import train_test_split

import tensorflow as tf
from numpy.random import seed
```

We create a dataset for binary classification:

```python
plt.figure(figsize=(4,4))
X, y = make_circles(1000, factor=0.3, noise=0.1)
mglearn.discrete_scatter(X[:,0],X[:,1], y)
```

Tensorflow needs the labels to be one-hot-encoded (i.e. encoded using dummy variables). For that, we can e.g. use Keras' "to_categorical"-function:

```python
print("y as a single binary variable:\n", y[0:10],'\n')
Y = to_categorical(y,num_classes=2)
print("One-hot-encoding of y:\n",Y[0:10])
```

As usual, the data is split into train, test and validation sets:

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, random_state=0)
```

We are now ready to train a neural network: 

```python
# If you want reproducable results, uncomment the following 4 lines:
seed(1)
tf.random.set_seed(2)

model = Sequential() # Network is going to be a sequence of layers - this kind of model is called "sequential"

model.add(Dense(units=64, activation='relu', input_dim=2)) # Each new line is a new layer! The first layer is called the input-layer
model.add(Dense(units=10, activation='tanh'))
model.add(Dense(units=2, activation='softmax')) # The last layer is called the output layer.

model.compile(loss='categorical_crossentropy', optimizer='sgd', metrics=['accuracy'])

model.fit(X_train, Y_train, epochs=500, batch_size=100, verbose=1)
```

As usual, we compare the accuracy on training and testing data in order to check if the model is overfitted:

```python
print("Accuracy on training data: {}".format(model.evaluate(X_train, Y_train)))
print("Accuracy on testing data: {}".format(model.evaluate(X_test, Y_test)))
```

Here is the decision boundary for the fully trained model:

```python
plt.figure(figsize=(4,4))
mglearn.discrete_scatter(X[:,0],X[:,1], Y[:,0])
mglearn.plots.plot_2d_separator(model, X, eps=.5, fill=True, alpha=0.3)
```
